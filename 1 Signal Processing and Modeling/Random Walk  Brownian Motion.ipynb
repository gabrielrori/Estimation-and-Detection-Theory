# Random walk & Brownian motion

```python
import numpy as np
from scipy.stats import binom
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Ruido blanco Bernoulli & Gaussiano
N = 1000
wnb = 2*np.random.randint(2,size=(N)) - 1
wng = np.random.randn(N)

plt.figure(figsize=[12,4])
plt.stem(wnb,basefmt="")
plt.grid('on')
plt.xlabel('$n$')
plt.xlim([0,100])
plt.ylim([-2.5,2.5])
plt.title('Ruido Blanco Bernoulli')

plt.figure(figsize=[12,4])
plt.stem(wng,basefmt="")
plt.grid('on')
plt.xlabel('$n$')
plt.xlim([0,100])
plt.ylim([-2.5,2.5])
plt.title('Ruido Blanco Gaussiano')

```

```python
# Random walk (camino aleatorio) y Brownian motion
rwn = np.concatenate((np.array([0]),np.cumsum(wnb)))
bmn = np.concatenate((np.array([0]),np.cumsum(wng)))

plt.figure(figsize=[12,4])
plt.stem(rwn,basefmt="")
plt.grid('on')
plt.xlabel('$n$')
plt.xlim([0,100])
plt.ylim([-20,20])
plt.title('Camino aleatorio')

plt.figure(figsize=[12,4])
plt.plot(bmn)
plt.grid('on')
plt.xlabel('$n$')
plt.xlim([0,100])
plt.ylim([-20,20])
plt.title('Movimientro Browniano')

```

```python
# Sample paths
N = 50
P = 10000

wnb2 = 2*np.random.randint(2,size=(N,P)) - 1
wng2 = np.random.randn(N,P)

rwn2 = np.concatenate((np.zeros((1,P)),np.cumsum(wnb2,axis=0)))
bmn2 = np.concatenate((np.zeros((1,P)),np.cumsum(wng2,axis=0)))

print('Random walk')
print('Mean:',np.mean(rwn2[N,:]))
print('Var:',np.var(rwn2[N,:]))

print('Brownian motion')
print('Mean:',np.mean(bmn2[N,:]))
print('Var:',np.var(bmn2[N,:]))

plt.figure(figsize=[8,4])
plt.plot(rwn2[:,0:10])
plt.grid('on')
plt.xlabel('$n$')
plt.xlim([0,N])
#plt.ylim([-20,20])
plt.title('Sample paths - Random walk')

plt.figure(figsize=[8,4])
plt.plot(bmn2[:,0:10])
plt.grid('on')
plt.xlabel('$n$')
plt.xlim([0,N])
#plt.ylim([-20,20])
plt.title('Sample paths - Brownian motion')
```
