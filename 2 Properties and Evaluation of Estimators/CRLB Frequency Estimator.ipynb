# CRLB Sinusoidal Frequency Estimator

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#plt.rcParams['font.size'] = '10'
```

Consider a sinusoidal signal in additive white Gaussian noise <br>
$x[n] = s[n;\theta]+w[n],\, n=0,1,\dots,N-1,$ and $w[n]\sim\mathcal{N}(0,\sigma^2)\,\forall n$<br>
where
$$ s[n,f_o] = A\cos(2 \pi f_o n + \phi), \quad 0<f_o<\tfrac{1}{2} $$
and amplitude $A$ and phase $\phi$ are known.

```python
N = 10
sigma = 4
A = 1
phi = 0*np.pi/2

fo = 0.5* np.arange(1,100)/100

sumfo = 0
for n in range(N):
    sumfo += (2*np.pi*n*np.sin(2*np.pi*fo*n+phi))**2

var_crlb = sigma**2 / (A**2*sumfo)
```

```python
plt.figure(figsize=[8,4])
plt.plot(fo,var_crlb)
plt.xlabel('Frequency')
plt.ylabel('CRLB')
plt.grid('on')
```
