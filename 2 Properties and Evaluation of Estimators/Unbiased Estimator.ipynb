# Unbiased Estimator

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
%matplotlib inline
#plt.rcParams['font.size'] = '10'
```

```python
A = 10
sigma = 2
N = 10
M = 10000

xu = A + sigma*np.random.randn(N,M)
xu = np.sum(xu,axis=0)/N

xb = (A-sigma) + sigma*np.random.rand(N,M)
xb = np.sum(xb,axis=0)/N
```

```python
plt.figure()
plt.plot(xu)
plt.plot(xb)
plt.xlim([0,100])
plt.xlabel('n')
plt.grid('on')
plt.legend(['Unbiased','Biased'])
```

```python
plt.figure()

plt.hist(xu,density=True)

a0 = np.linspace(A - 3.5*sigma, A + 3.5*sigma, 100)
plt.plot(a0, stats.norm.pdf(a0, A, sigma),'g')

plt.plot(a0, stats.norm.pdf(a0, A, sigma/np.sqrt(N)),'r')

plt.xlabel('Parameter A')
plt.title('Unbiased Estimator')
plt.grid('on')


plt.figure()

plt.hist(xb,density=True)

a0 = np.linspace(A - 3.5*sigma, A + 3.5*sigma, 100)
plt.plot(a0, stats.norm.pdf(a0, A, sigma),'g')

plt.plot(a0, stats.uniform.pdf(a0, A-sigma/2 - sigma/np.sqrt(N)/2 , sigma/np.sqrt(N)),'r')

plt.xlabel('Parameter A')
plt.title('Biased Estimator')
plt.grid('on')
```
